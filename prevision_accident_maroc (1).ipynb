```python
dataset=dataset[['ANN_ACC','mois','jour','COD_PRV','COD_LUM','COD_INT','feries']]
```

```python
dataset=dataset.groupby(['ANN_ACC','mois','jour','COD_PRV','COD_LUM','COD_INT','feries']).size().reset_index(name='counts')
dataset
```

```python
import pandas as pd

df = pd.DataFrame({'feries': ['False', 'True']})
dataset[['feries_False', 'feries_True']]=pd.get_dummies(dataset['feries'],prefix='feries')

df = pd.DataFrame({'LUM': [2,  3, 10]})
dataset[['LUM_2', 'LUM_3', 'LUM_10']]=pd.get_dummies(dataset['COD_LUM'],prefix='LUM')

df = pd.DataFrame({'COD_INT': [ 2,  3,  5, 10,  6,  4]})
dataset[['COD_INT_2', 'COD_INT_3', 'COD_INT_5',
         'COD_INT_10', 'COD_INT_6', 'COD_INT_4']]=pd.get_dummies(dataset['COD_INT'],prefix='COD_INT')

def anne_accident(ANN_ACC):
    return ANN_ACC-2010

#dataset['ANN_ACC_transf']=da taset['ANN_ACC'].apply(anne_accident)


dataset_train=dataset[dataset['ANN_ACC']!=2014]
dataset_test=dataset[dataset['ANN_ACC']==2014]
```

```python
X_train = dataset_train.drop(['COD_LUM','COD_INT','counts','feries'], axis=1)
y_train = dataset_train['counts']

X_test = dataset_test.drop(['COD_LUM','COD_INT','counts','feries'], axis=1)
y_test = dataset_test['counts']
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_val_predict, train_test_split
```

```python
dtm = DecisionTreeRegressor(max_depth=25,
                           min_samples_split=2,
                           max_leaf_nodes=2000)
X
x=dtm.fit(X_train,y_train)
print("R-Squared on train dataset={}".format(dtm.score(X_test,y_test)))

dtm.fit(X_test,y_test)   
print("R-Squaredon test dataset={}".format(dtm.score(X_test,y_test)))
```

```python
from sklearn.externals import joblib
joblib.dump(dtm, open("model11.sav","wb"))
```
